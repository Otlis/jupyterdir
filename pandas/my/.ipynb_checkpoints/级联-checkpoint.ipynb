```python
import numpy as np
import pandas as pd
from pandas import Series,DataFrame
```

```python
df1 = DataFrame(np.random.randint(0,150,size = (5,3)),columns=['Python','Math','En'],index = list('ABCDE'))

df2 = DataFrame(np.random.randint(0,150,size = (5,3)),columns=['Python','Math','En'],index=list('HIJKL'))

display(df1,df2)
```

```python
pd.concat([df1,df2])
```

```python
df3 = DataFrame(np.random.randint(0,150,size = (5,3)),columns=['Chinese','Physic','Chem'],index = list('QWRTY'))

df3
```

```python
# 级联时有些样本没有一些属性，那么级联出来的就是nan
df4=pd.concat([df1,df3],axis=1)
# 然后填充空值
df4.fillna(0)
```

```python
df5 = DataFrame(np.random.randint(0,150,size = (5,3)),columns=['Python','Math','Java'],index = list('ABCGH'))
df5
```

```python
pd.concat([df1,df5],join='outer')
```

```python
# 忽略掉原来的index
pd.concat([df1,df5],ignore_index=True)
```

```python
# 期末考试
df6 =DataFrame(np.random.randint(0,150,size = (5,3)),columns=df1.columns,index= df1.index)
df6
```

```python
df7=pd.concat([df1,df6],keys=['期中','期末'])
df7
```

```python
df7.unstack(level=0)
```

```python
df7.unstack(level=0).stack()
```

```python
df1 = DataFrame({'id':[1,10,1024],'name':['po','alisa','Michael'],'sex':['male','female','male']})
df1
```

```python
df2 = DataFrame({'id':[1,11,1025],'salary':[10000,20000,30000],'age':[20,30,40]})
df2
```

```python
pd.concat([df1,df2])
```

```python
df1.merge(df2)
```

```python
df2 = DataFrame({'id':[1,1,1],'salary':[10000,20000,30000],'age':[20,30,40]})
df2
```

```python
df1.merge(df2)
```

```python
df3 = DataFrame({'id':[1,1,1],'name':['po','alisa','Michael'],'sex':['male','female','male']})
df3
```

```python
df2.merge(df3)
```

```python
# 就相当于  table1 join table2 on t1.name=t2.name
pd.merge(df1,df2,on='name')
```

```python
# 这就是左连接  table1.sid=table2.id连接条件
pd.merge(df1,df2,how='left',left_on='sid',right_on='id')
```

```python
df = DataFrame(np.random.randint(0,150,size = (10,3)),columns=['Python','Math','En'],index = list('ABCDEFHIJK'))
df
```

```python
s=df.sum(axis=1)
df2=DataFrame(s,columns=['sum_score'])
df2
```

```python
pd.merge(df,df2,left_index=True,right_index=True)
```
