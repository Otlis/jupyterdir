```python
import numpy as np
import pandas as pd
from pandas import Series,DataFrame
```

```python
areas = pd.read_csv('./state-areas.csv')
areas
```

```python
abbr = pd.read_csv('./state-abbrevs.csv')
abbr
```

```python
pop = pd.read_csv('./state-population.csv')
pop
```

```python
pop.head()
```

```python
pop2=pd.merge(abbr,pop,left_on='abbreviation',right_on='state/region',how='outer')
```

```python
pop2.head()
```

```python
pop2.drop(labels='abbreviation',axis=1,inplace=True)
pop2.head()
```

```python
pop2.isnull().any()
```

```python
cond=pop2['state'].isnull()
a=pop2['state/region'][cond]
a.unique()
```

```python
cond=pop2['state/region']=='PR'
pop2['state'][cond]='Puerto Rico'
```

```python
cond = pop2['state/region'] == 'USA'

pop2['state'][cond] = 'United States'
```

```python
pop2.isnull().any()
```

```python
# inplace=True的意思是替代之前的数据，不会在内存中生成一个新的df
pop2.dropna(inplace=True)
```

```python
pop2.isnull().any()
```

```python
pop2.head()
```

```python
areas.head()
```

```python
pop3 = pd.merge(pop2,areas,how='outer',left_on='state',right_on='state')
display(pop3.head(),pop3.shape)
```

```python
pop3.isnull().any()
```

```python
cond=pop3['area (sq. mi)'].isnull()
pop3[cond]
```

```python
areas['area (sq. mi)'].sum()
```

```python
pop3.fillna(3790399,inplace=True)
```

```python
pop3.notnull().all()
```

```python
pop3.head()
```

找出2010年的全民人口数据,df.query(查询语句)

```python
pop_2010=pop3.query("year=='2010' and ages=='total'")
pop_2010.set_index('state/region',inplace=True)
pop_2010.head()
```

```python
s=pop_2010['population']/pop_2010['area (sq. mi)']
s
```

```python
pop_density=DataFrame(s,columns=['pop_density'])
pop_density.round(2)
```

```python
# on 连接条件
result=pd.merge(pop_2010,pop_density,on='state/region')
```

```python
result.sort_values(by="pop_density",inplace=True)
```
