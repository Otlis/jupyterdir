```python
df=DataFrame({'item':np.random.randint(0,10,size=100),
          'sailer':np.random.randint(0,3,size=100),
           'weight':np.random.randint(30,300,size=100),
           'price':np.random.randint(2,20,size=100)
          })
df         
```

```python
def convert_item(x):
    if x<3:
        return '黄瓜'
    elif x<6:
        return '香蕉'
    elif x<8:
        return '火龙果'
    else:
        return '苹果'

df['item']=df['item'].map(convert_item)
```

```python
def convert_sailer(x):
    if x==0:
        return '张大爷'
    if x==1:
        return '李大爷'
    if x==2:
        return '孙大爷'
df['sailer']=df['sailer'].map(convert_sailer)
```

```python
def convert_weight(x):
    if x<100:
        return 50
    elif x <200:
        return 100
    else:
        return 150
df['weight']=df['weight'].map(convert_weight)
```

```python
df
```

```python
# 统计每个人销售了多少斤？
ret=df.groupby(['sailer'])['weight']
```

```python
ret.sum()
```

```python
# 调用apply 传递一个方法
ret.apply(np.sum)
```

```python
def my_fun(x):
    return (x.size,x.sum(),np.round(x.mean(),decimals=2))
ret.apply(my_fun)
```

```python
g=df.groupby(['sailer','item'])
g
```

```python
# 价格求平均，销量求和

def my_fun(x):
    return(x['price'].mean(),x['weight'].sum())
g.apply(my_fun)
```

```python
# agg 方法里面传递的也是一个函数，类似于apply
g.agg(['mean','sum'])
```

```python
g.agg({'weight':'sum','price':'mean'}).round(2)
```
